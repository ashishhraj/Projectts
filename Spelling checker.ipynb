```python
import tkinter as tk
from tkinter.scrolledtext import ScrolledText
from spellchecker import SpellChecker  # Import the SpellChecker class
import nltk

class SpellingChecker:
    def __init__(self):
        self.root = tk.Tk()
        self.root.geometry("600x500")  #Create a Tkinter window (self.root) with a size of 600x500 pixels.

        self.text = ScrolledText(self.root, font=("Arial", 16))
        self.text.bind("<KeyRelease>", self.check)  #the "<KeyRelease>" event to the self.check method.
                                                   #This event triggers whenever a key is released while typing in the text box.
        self.text.pack()    #Pack the self.text widget to display it in the window.

        self.spell = SpellChecker()  # Create a SpellChecker instance

        self.old_spaces = 0  #Initialize a variable self.old_spaces to keep track of the number of spaces in the text box.

        self.root.mainloop()

    def check(self, event):  #The check method is called whenever there is a key release event. 
        content = self.text.get("1.0", tk.END)
        space_count = content.count(" ")

        if space_count != self.old_spaces:
            self.old_spaces = space_count

            for tag in self.text.tag_names():
                if tag != "sel":       #the "sel" tag is used to represent the selected text within the tkinter text widget.
                    self.text.tag_delete(tag)

            words = content.split(" ")
            for i, word in enumerate(words):
                # Check if the lowercase version of the word is incorrect
                corrected_word = self.spell.correction(word.lower()) 
                
                if corrected_word != word.lower():
                    position = content.find(word)
                    tag_name = f"error_{position}"  # Create a unique tag name
                    self.text.tag_add(tag_name, f"1.{position}", f"1.{position + len(word)}")
                    self.text.tag_config(tag_name, foreground="red")
                    
                    # Replace the incorrect word with the corrected word
                    words[i] = corrected_word

            # Update the content of the text widget with corrected words
            corrected_content = " ".join(words)
            self.text.delete("1.0", tk.END)
            self.text.insert(tk.END, corrected_content)

if __name__ == "__main__":
    SpellingChecker()
```
